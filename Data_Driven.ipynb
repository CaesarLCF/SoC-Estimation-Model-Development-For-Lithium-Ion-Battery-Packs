### **3). Process the Datasets:**

```python
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.metrics import mean_squared_error, mean_absolute_error, mean_absolute_percentage_error, r2_score
from tqdm import tqdm
from torch.utils.data import TensorDataset, DataLoader
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
import numpy as np
import matplotlib.pyplot as plt

# Set x = Input, y = Output
x = df.drop(columns=['SoC']).values  # Features
y = df['SoC'].values  # Labels

# Set up the MinMaxScaler
x_scaler = MinMaxScaler()
y_scaler = MinMaxScaler()

# Scale the x and y with MinMaxScaler
x_scaled = x_scaler.fit_transform(x)
y_scaled = y_scaler.fit_transform(y.reshape(-1, 1))

# Split the data into training, validation, and test sets
x_train, x_rem, y_train, y_rem = train_test_split(x_scaled, y_scaled, train_size=0.8, random_state=42)
x_val, x_test, y_val, y_test = train_test_split(x_rem, y_rem, test_size=0.5, random_state=42)

# Chnage the data into tensor format
x_train_tensor = torch.FloatTensor(x_train)
y_train_tensor = torch.FloatTensor(y_train).view(-1, 1)

x_val_tensor = torch.FloatTensor(x_val)
y_val_tensor = torch.FloatTensor(y_val).view(-1, 1)

x_test_tensor = torch.FloatTensor(x_test)
y_test_tensor = torch.FloatTensor(y_test).view(-1, 1)

train_dataset = TensorDataset(x_train_tensor, y_train_tensor)
val_dataset = TensorDataset(x_val_tensor, y_val_tensor)
test_dataset = TensorDataset(x_test_tensor, y_test_tensor)

# Set up the Hyperparameter = batch_size
train_loader = DataLoader(train_dataset, batch_size=96, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=96, shuffle=False)
test_loader = DataLoader(test_dataset, batch_size=96, shuffle=False)
```

# **Backpropagation Neural Network (BPNN)**

---



### **a). Train the BPNN model:**

```python
# Define the BPNN model
class BPNN(nn.Module):
    def __init__(self):
        super(BPNN, self).__init__()
        self.flatten = nn.Flatten()
        self.model = nn.Sequential(
            nn.Linear(3, 512), # 3 input features + Hidden 1
            nn.ReLU(),
            nn.Linear(512, 512), # Hidden 2
            nn.ReLU(),
            nn.Linear(512, 1024), # Hidden 3
            nn.ReLU(),
            nn.Linear(1024, 1024), # Hidden 4
            nn.ReLU(),
            nn.Linear(1024, 1024), # Hidden 5
            nn.ReLU(),
            nn.Linear(1024, 1024), # Hidden 6
            nn.ReLU(),
            nn.Linear(1024, 1024), # Hidden 7
            nn.ReLU(),
            nn.Linear(1024, 1) # Output Layer
        )

    def forward(self, x):
        x = self.flatten(x)
        out = self.model(x)
        return out

# Function to save the best model
def save_best_model(epoch, loss, model, optimizer, path):
    torch.save({
        'epoch': epoch,
        'model_state_dict': model.state_dict(),
        'optimizer_state_dict': optimizer.state_dict(),
        'loss': loss
    }, path)
    print(f"Saved best model with validation loss {loss:.6f}")
    print('===================================================================')

# Set random seed for reproducibility
torch.manual_seed(999)

# Initialize model, loss function, and optimizer
device = 'cuda:0' if torch.cuda.is_available() else 'cpu'
model = BPNN().to(device)
loss_fn = torch.nn.MSELoss() # Hyperparameter = Loss Function
optimizer = optim.Adam(model.parameters(), lr=0.0001) # Hyperparameter = Learning Rate / Optimizer

# Initialize variables to track the best model
best_val_loss = float("inf")  # Pre-set the best validation loss value (ensure your newly trained model is better than last trined model)
model_save_path = '/content/drive/MyDrive/BPNN.pth' # Location to save the best model

# Training loop
num_epochs = 30 # Hyperparameter = Epoch
for epoch in range(num_epochs):
    model.train()

    # Training with progress bar
    train_loss = 0
    with tqdm(total=len(train_loader), desc=f'Epoch {epoch+1}/{num_epochs}', unit='batch') as pbar:
        for X_batch, y_batch in train_loader:
            X_batch, y_batch = X_batch.to(device), y_batch.to(device)
            optimizer.zero_grad()
            outputs = model(X_batch)
            loss = loss_fn(outputs, y_batch)
            loss.backward()
            optimizer.step()

            train_loss += loss.item()
            pbar.set_postfix({'Loss': train_loss / (pbar.n + 1)})
            pbar.update(1)

    # Validation with progress bar
    model.eval()
    val_loss = 0
    with torch.no_grad():
        with tqdm(total=len(val_loader), desc='Validation', unit='batch') as pbar:
            for X_batch, y_batch in val_loader:
                X_batch, y_batch = X_batch.to(device), y_batch.to(device)
                outputs = model(X_batch)
                loss = loss_fn(outputs, y_batch)
                val_loss += loss.item()
                pbar.update(1)

    val_loss = val_loss / len(val_loader)
    print(f'Epoch {epoch+1}/{num_epochs}, Train Loss: {train_loss / len(train_loader)}, Validation Loss: {val_loss}')

    # Save the model if validation loss improves
    if val_loss < best_val_loss:
        best_val_loss = val_loss
        save_best_model(epoch, val_loss, model, optimizer, model_save_path)
```

### **b). Evaluate the BPNN model:**

```python
# Define the BPNN model
class BPNN(nn.Module):
    def __init__(self):
        super(BPNN, self).__init__()
        self.flatten = nn.Flatten()
        self.model = nn.Sequential(
            nn.Linear(3, 512), # 3 input features + Hidden 1
            nn.ReLU(),
            nn.Linear(512, 512), # Hidden 2
            nn.ReLU(),
            nn.Linear(512, 1024), # Hidden 3
            nn.ReLU(),
            nn.Linear(1024, 1024), # Hidden 4
            nn.ReLU(),
            nn.Linear(1024, 1024), # Hidden 5
            nn.ReLU(),
            nn.Linear(1024, 1024), # Hidden 6
            nn.ReLU(),
            nn.Linear(1024, 1024), # Hidden 7
            nn.ReLU(),
            nn.Linear(1024, 1) # Output Layer
        )

    def forward(self, x):
        x = self.flatten(x)
        out = self.model(x)
        return out

# Initialize the model
device = 'cuda:0' if torch.cuda.is_available() else 'cpu'
model = BPNN().to(device)  # Ensure the model is defined before loading
loss_fn = torch.nn.MSELoss()  # Hyperparameter = Loss Function

# Load the saved model state
model_save_path = '/content/drive/MyDrive/BPNN.pth' # Location to load the best model
checkpoint = torch.load(model_save_path, map_location=device)
model.load_state_dict(checkpoint['model_state_dict'])

# Evaluation of BPNN Model: Test Data
model.eval()
with torch.no_grad():
    test_loss = 0
    y_preds = []
    with tqdm(total=len(test_loader), desc='Testing', unit='batch') as pbar:
        for X_batch, y_batch in test_loader:
            X_batch, y_batch = X_batch.to(device), y_batch.to(device)
            outputs = model(X_batch)  # Remove unsqueeze if not needed
            loss = loss_fn(outputs, y_batch)
            test_loss += loss.item()
            y_preds.append(outputs.cpu().numpy())
            pbar.update(1)

    y_pred = np.concatenate(y_preds, axis=0)
    y_test_orig = y_scaler.inverse_transform(y_test)
    y_pred_orig = y_scaler.inverse_transform(y_pred)

    mse = mean_squared_error(y_test_orig, y_pred_orig)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y_test_orig, y_pred_orig)
    mape = mean_absolute_percentage_error(y_test_orig, y_pred_orig)
    r2 = r2_score(y_test_orig, y_pred_orig)

    print(f'Test Loss: {test_loss / len(test_loader)}')
    print(f'MSE: {mse:.6f}')
    print(f'RMSE: {rmse:.6f}')
    print(f'MAE: {mae:.6f}')
    print(f'MAPE: {mape:.6f}')
    print(f'R²: {r2:.6f}')

# Plotting the first 100 samples
plt.figure(figsize=(15, 7))
plt.plot(y_test_orig[:100], label='Actual SoC', color='blue')
plt.plot(y_pred_orig[:100], label='Predicted SoC', color='red', linestyle='dashed')
plt.xlabel('Sample Index')
plt.ylabel('SoC')
plt.title(f'Actual vs. Predicted SoC (First 100 Samples)')
plt.legend()
plt.grid(True)
plt.show()

# Plotting the first 10 samples
plt.figure(figsize=(15, 7))
plt.plot(y_test_orig[:10], label='Actual SoC', color='blue')
plt.plot(y_pred_orig[:10], label='Predicted SoC', color='red', linestyle='dashed')
plt.xlabel('Sample Index')
plt.ylabel('SoC')
plt.title(f'Actual vs. Predicted SoC (First 10 Samples)')
plt.legend()
plt.grid(True)
plt.show()
```

# **Long Short-Term Memory (LSTM)**
---



### **a). Train the LSTM model:**

```python
# Define the LSTM model
class LSTM(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers, output_size):
        super(LSTM, self).__init__()
        self.output_size = output_size
        self.num_layers = num_layers
        self.input_size = input_size
        self.hidden_size = hidden_size

        self.lstm = nn.LSTM(input_size=input_size, hidden_size=hidden_size, num_layers=num_layers, batch_first=True)
        self.fc_1 = nn.Linear(hidden_size, 64) # Hidden 1
        self.fc_2 = nn.Linear(64, 128) # Hidden 2
        self.fc_3 = nn.Linear(128, 256) # Hidden 3
        self.fc_4 = nn.Linear(256, 512) # Hidden 4
        self.fc_5 = nn.Linear(512, 512) # Hidden 5
        self.fc_6 = nn.Linear(512, 1024) # Hidden 6
        self.fc_7 = nn.Linear(1024, 1024) # Hidden 7
        self.outp = nn.Linear(1024, output_size)
        self.relu = nn.ReLU()

    def forward(self,x):
        h_0 = torch.zeros(self.num_layers, x.size(0), self.hidden_size).to(device)
        c_0 = torch.zeros(self.num_layers, x.size(0), self.hidden_size).to(device)

        out, (hn, cn) = self.lstm(x, (h_0, c_0))

        # Take the last output for each sequence in the batch
        out = out[:, -1, :]  # This will give a tensor of shape (batch_size, hidden_size)

        out = self.relu(out)
        out = self.fc_1(out)
        out = self.relu(out)
        out = self.fc_2(out)
        out = self.relu(out)
        out = self.fc_3(out)
        out = self.relu(out)
        out = self.fc_4(out)
        out = self.relu(out)
        out = self.fc_5(out)
        out = self.relu(out)
        out = self.fc_6(out)
        out = self.relu(out)
        out = self.fc_7(out)
        out = self.relu(out)
        out = self.outp(out)
        return out

# Function to save the best model
def save_best_model(epoch, loss, model, optimizer, path):
    torch.save({
        'epoch': epoch,
        'model_state_dict': model.state_dict(),
        'optimizer_state_dict': optimizer.state_dict(),
        'loss': loss
    }, path)
    print(f"Saved best model with validation loss {loss:.6f}")
    print('===================================================================')

# Set random seed for reproducibility
torch.manual_seed(999)

# Initialize model, loss function, and optimizer
device = 'cuda:0' if torch.cuda.is_available() else 'cpu'
input_size = 3
hidden_size = 20
num_layers = 2
output_size = 1
model = LSTM(input_size, hidden_size, num_layers, output_size).to(device)
loss_fn = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.0005)

# Initialize variables to track the best model
best_val_loss = float("inf")
model_save_path = '/content/drive/MyDrive/LSTM.pth'

# Training loop
num_epochs = 30
for epoch in range(num_epochs):
    model.train()
    train_loss = 0

    # Training with progress bar
    with tqdm(total=len(train_loader), desc=f'Epoch {epoch+1}/{num_epochs}', unit='batch') as pbar:
        for X_batch, y_batch in train_loader:
            X_batch, y_batch = X_batch.to(device), y_batch.to(device)
            optimizer.zero_grad()
            outputs = model(X_batch.unsqueeze(1))  # Adding a dimension for batch_first=True
            loss = loss_fn(outputs, y_batch)
            loss.backward()
            optimizer.step()
            train_loss += loss.item()
            pbar.set_postfix({'Loss': train_loss / (pbar.n + 1)})
            pbar.update(1)

    # Validation
    model.eval()
    val_loss = 0
    with torch.no_grad():
        for X_batch, y_batch in val_loader:
            X_batch, y_batch = X_batch.to(device), y_batch.to(device)
            outputs = model(X_batch.unsqueeze(1))
            loss = loss_fn(outputs, y_batch)
            val_loss += loss.item()

    val_loss = val_loss / len(val_loader)
    print(f'Epoch {epoch+1}/{num_epochs}, Train Loss: {train_loss / len(train_loader)}, Validation Loss: {val_loss}')

    # Save the model if validation loss improves
    if val_loss < best_val_loss:
        best_val_loss = val_loss
        save_best_model(epoch, val_loss, model, optimizer, model_save_path)
```

```python
# Define the CNN model
class CNN(nn.Module):
    def __init__(self):
        super().__init__()
        self.flatten = nn.Flatten()
        self.conv1 = nn.Conv1d(in_channels=3, out_channels=10, kernel_size=1)
        self.pool = nn.MaxPool1d(kernel_size=1)

        # Calculate the size of the output after conv1 and pool layers
        self.fc1 = nn.Linear(10, 128)
        self.fc2 = nn.Linear(128, 128)
        self.fc3 = nn.Linear(128, 256)
        self.fc4 = nn.Linear(256, 256)
        self.fc5 = nn.Linear(256, 512)
        self.fc6 = nn.Linear(512, 512)
        self.fc7 = nn.Linear(512, 512)
        self.fc8 = nn.Linear(512, 512)
        self.fc9 = nn.Linear(512, 1)
        self.relu = nn.ReLU()

    def forward(self, x):
        x = x.unsqueeze(2)
        out = self.conv1(x)
        out = self.relu(out)
        out = self.pool(out)
        out = self.flatten(out)
        out = self.fc1(out)
        out = self.relu(out)
        out = self.fc2(out)
        out = self.relu(out)
        out = self.fc3(out)
        out = self.relu(out)
        out = self.fc4(out)
        out = self.relu(out)
        out = self.fc5(out)
        out = self.relu(out)
        out = self.fc6(out)
        out = self.relu(out)
        out = self.fc7(out)
        out = self.relu(out)
        out = self.fc8(out)
        out = self.relu(out)
        out = self.fc9(out)
        return out  # <-- Add this line to return the output

# Initialize the model
device = 'cuda:0' if torch.cuda.is_available() else 'cpu'
model = CNN().to(device)
loss_fn = nn.MSELoss()

# Load the saved model state
model_save_path = '/content/drive/MyDrive/CNN.pth'
checkpoint = torch.load(model_save_path, map_location=device)
model.load_state_dict(checkpoint['model_state_dict'])

# Evaluation of BPNN Model:
model.eval()
with torch.no_grad():
    test_loss = 0
    y_preds = []
    with tqdm(total=len(test_loader), desc='Testing', unit='batch') as pbar:
        for X_batch, y_batch in test_loader:
            X_batch, y_batch = X_batch.to(device), y_batch.to(device)
            outputs = model(X_batch)
            loss = loss_fn(outputs, y_batch)
            test_loss += loss.item()
            y_preds.append(outputs.cpu().numpy())
            pbar.update(1)

    y_pred = np.concatenate(y_preds, axis=0)
    y_test_orig = y_scaler.inverse_transform(y_test)
    y_pred_orig = y_scaler.inverse_transform(y_pred)

    mse = mean_squared_error(y_test_orig, y_pred_orig)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y_test_orig, y_pred_orig)
    mape = mean_absolute_percentage_error(y_test_orig, y_pred_orig)
    r2 = r2_score(y_test_orig, y_pred_orig)

    print(f'Test Loss: {test_loss / len(test_loader)}')
    print(f'MSE: {mse:.6f}')
    print(f'RMSE: {rmse:.6f}')
    print(f'MAE: {mae:.6f}')
    print(f'MAPE: {mape:.6f}')
    print(f'R²: {r2:.6f}')

# Plotting the first 100 samples
plt.figure(figsize=(15, 7))
plt.plot(y_test_orig[:100], label='Actual SoC', color='blue')
plt.plot(y_pred_orig[:100], label='Predicted SoC', color='red', linestyle='dashed')
plt.xlabel('Sample Index')
plt.ylabel('SoC')
plt.title(f'Actual vs. Predicted SoC (First 100 Samples)')
plt.legend()
plt.grid(True)
plt.show()

# Plotting the first 10 samples
plt.figure(figsize=(15, 7))
plt.plot(y_test_orig[:10], label='Actual SoC', color='blue')
plt.plot(y_pred_orig[:10], label='Predicted SoC', color='red', linestyle='dashed')
plt.xlabel('Sample Index')
plt.ylabel('SoC')
plt.title(f'Actual vs. Predicted SoC (First 10 Samples)')
plt.legend()
plt.grid(True)
plt.show()
```
